```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\dd\Documents\Google_Stock_Price_Test.csv")
print(df)
df['Date'] = pd.to_datetime(df['Date'])
start_date = '2017-01-01'
end_date = '2017-01-31'
filtered_df = df[(df['Date']>= start_date) & (df['Date']<= end_date)]
dates = filtered_df['Date']
opening_prices = filtered_df['Open']
closing_prices = filtered_df['Close']
plt.plot(dates,opening_prices, label ='Opening Price')
plt.plot(dates,closing_prices,label='Closing Price')
plt.title('Stock Prices(Opening and Closing)')
plt.xlabel('Data')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# write a program to find factorial using function

```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n - 1)
num = int(input("Enter a number: "))
if num<0:
    print("Factorial is not defined for negative numbers.")
else:
    result = factorial(num)
    print("The factorial of", num, "is", result)
```
